```python
import numpy as np
from materialLibs import *
from unitsLibs import *
```

```python
# 1. Definir especies y propiedades
species = ["CO2", "N2"]
WM, mu, Dm, cpg_molar, cpg_mass, k, H = get_properties_gas(species)
propiedades_gas = {
    "species": species,
    "MW": WM,
    "mu": mu,
    "Dm": Dm,
    "Cp_molar": cpg_molar,
    "Cp_mass": cpg_mass,
    "k": k,
    "H": H}
```

```python
# LLENADO DE UN TANQUE PARA DIFERENTES CONFIGURACIONES DE VALVULAS
# 1. Crear y definir las valvulas
valve_inlet_config = {
    "Name": "Entrada",
    "Cv_max": 450,
    "valve_type": "linear",
    "logic": "linear",
    "logic_params": {"start": 0, "duration": 18},
    "opening_direction": "co"
}

valve_outlet_config = {
    "Name": "Salida",
    "Cv_max": 0.0,  # cerrada
    "valve_type": "linear",
    "logic": "step",
    "logic_params": {"start": 0, "duration": 2},
    "opening_direction": "co"
}

tank = Tank(
    Name="Tanque de prueba",
    Longitud=2,     # m
    Diametro=.61,     # m
    Espesor=0.02,     # m
    prop_gas=propiedades_gas,
    Valve_inlet_config=valve_inlet_config,
    Valve_outlet_config=valve_outlet_config
)

# 4. Condiciones iniciales (vacío)
tank.initialC_info(P0=3.5e5,T0=298.0,x0=[0.5, 0.5])
# 5. Condiciones de frontera (alimentación)
tank.boundaryC_info(Pin=5e5,Tin=350.0,xin=[0.2, 0.8],Pout=1e5)
# 6. Thermal_info
tank.thermal_info(adi=True,kw=1e99,hint=1e99,hext=1e99,Tamb=298.15)
# 7. Parámetros de simulación
tank.simulation_info(saveData=1,endTime=20)
```

# LLENADO DE TANQUE

## Apertura lineal con logica lineal

```python
valve_inlet_config ={
    "Name": "linear-linear-CO",                # Nombre identificativo de la válvula (no funcional, solo para seguimiento)
    "Cv_max": 450,                             # Coeficiente máximo de caudal [Nm³/h·bar^0.5]
    "valve_type": "linear",                    # Tipo de válvula: la apertura se traduce en Cv de forma lineal (Cv = a * Cv_max)
    "logic": "linear",                         # Lógica de control: la apertura crece linealmente desde 0 hasta 1 (100 %)
    "logic_params": {
        "start": 0,                            # Momento en que empieza a abrir la válvula [s]
        "duration": 18                         # Tiempo que tarda en alcanzar apertura completa [s]
    },
    "opening_direction": "co"                  # Dirección de apertura: 'co' significa que se cierra al abrir (va de 1 a 0)
}

tank._initialize()
tank._solveTank()
tank._plotAll()

#tank._valve_inlet._plot()
#tank._plot()
#tank._valve_outlet._plot()
```

## Apertura lineal con logica escalonada

```python
config_new_inlet = {
    "Name": "Linial-Step-OC",                   # Nombre de la válvula (descriptivo, sin efecto en lógica)
    "Cv_max": 450.0,                            # Coeficiente máximo de caudal [Nm³/h·bar^0.5]
    "valve_type": "linear",                     # Tipo de válvula: Cv aumenta proporcionalmente con la apertura (lineal)
    "logic": "step",                            # Lógica: apertura escalonada en pasos discretos
    "logic_params": {
        "start": 0,                             # Inicio de la lógica de apertura [s]
        "duration": 20,                         # Duración total hasta alcanzar el valor máximo de apertura [s]
        "steps": 10                             # Número total de pasos (cada paso aumenta/disminuye 10 %)
    },
    "opening_direction": "co"                   # Dirección 'co' (close-to-open): empieza completamente abierta (100 %) y se va cerrando escalonadamente
}

tank._valve_inlet._update(config_new_inlet)
tank._initialize()
tank._solveTank()
tank._plotAll()
```

## Apertura lineal con logica senoidal

```python
config_new_inlet = {
    "Name": "Lineal-Sin-OC",                    # Nombre descriptivo de la válvula
    "Cv_max": 450.0,                             # Coeficiente máximo de caudal [Nm³/h·bar^0.5]
    "valve_type": "linear",                      # Tipo de válvula: Cv varía linealmente con la apertura (Cv = a * Cv_max)
    "logic": "sin",                              # Lógica: variación sinusoidal de la apertura
    "logic_params": {
        "start": 0,                              # Tiempo inicial de acción [s]
        "duration": 20,                          # Tiempo total del ciclo de apertura [s]
        "freq": 5                                # Número de ciclos completos (oscilaciones) dentro del intervalo de duración
    },
    "opening_direction": "co"                   # Dirección 'co' (close-to-open): se interpreta como inverso del perfil sinusoidal
}

tank._valve_inlet._update(config_new_inlet)
tank._initialize()
tank._solveTank()
tank._plotAll()
```

## Apertura lineal con logica sigmoidal

```python
config_new_inlet = {
    "Name": "Lineal-Sigmoid-OC",                # Nombre descriptivo de la válvula
    "Cv_max": 450.0,                             # Coeficiente máximo de caudal [Nm³/h·bar^0.5]
    "valve_type": "linear",                      # Tipo de válvula: Cv escala linealmente con la apertura (Cv = a * Cv_max)
    "logic": "sigmoid",                          # Lógica de control: curva sigmoide (suave al inicio y al final, más rápida al centro)
    "logic_params": {
        "start": 0,                              # Tiempo en que empieza a ejecutarse la lógica [s]
        "duration": 20,                          # Tiempo total para completar el perfil de apertura [s]
        "k": 10,                                 # Pendiente o "agudeza" de la curva (más alto = más brusco)
        "x0": 0.5                                # Punto de inflexión (normalizado entre 0 y 1 → a la mitad del tiempo)
    },
    "opening_direction": "co"                   # Dirección 'co' (close-to-open): se invierte la curva (empieza abierta y cierra suavemente)
}

tank._valve_inlet._update(config_new_inlet)
tank._initialize()
tank._solveTank()
tank._plotAll()
```

## Apertura lineal con logica polinómica

```python
config_new_inlet = {
    "Name": "Custom-Poly",                       # Nombre de la válvula (identificativo para gráficos o seguimiento)
    "Cv_max": 800.0,                             # Caudal máximo en condiciones normales [Nm³/h·bar^0.5]
    "valve_type": "linear",                      # Relación lineal entre apertura y coeficiente Cv (Cv = a * Cv_max)
    "logic": "poly",                             # Lógica: apertura basada en una función polinómica (potencia)
    "logic_params": {
        "start": 0,                              # Momento inicial del ciclo de apertura [s]
        "duration": 20,                          # Tiempo total hasta alcanzar apertura completa [s]
        "degree": 3                              # Grado del polinomio: cuanto mayor, más lenta la apertura al inicio
    },
    "opening_direction": "co"                    # Dirección: 'close-to-open' → comienza abierta (100 %) y se cierra
}

tank._valve_inlet._update(config_new_inlet)
tank._initialize()
tank._solveTank()
tank._plotAll()
```

```python
config_new_inlet =  {
    "Name": "Lineal_Rampa_CO",
    "Cv_max": 450.,                      # Coeficiente máximo de caudal
    "valve_type": "linear",             # Tipo de válvula (respuesta lineal)
    "logic": "ramp",                    # Lógica: rampa lineal de apertura
    "logic_params": {
        "start": 5,                     # Tiempo inicial de apertura [s]
        "duration": 10,                  # Duración total de la apertura [s]
        "height": 1                     # Altura máxima de apertura (1 = 100 %)
    },
    "opening_direction": "co"           # Se abre (open-to-close)
}

tank._valve_inlet._update(config_new_inlet)
tank._initialize()
tank._solveTank()
tank._plotAll()
```
