```python
 def isotherm(IsoFuncs, y, P_tot_Pa, T, method="IAST", Lamb=None, C=0.0, drop_thresh=0.1):
    iso_fun = IsoFuncs['iso_fun']
    species = IsoFuncs['species']

    y = np.array(y, dtype=float)
    N_total = len(y)
    Ptot_kPa = P_tot_Pa / 1000.0

    # --- detectar especies relevantes ---
    idx_keep = np.where(y >= drop_thresh)[0]
    idx_drop = np.where(y < drop_thresh)[0]

    # Si no queda nada relevante, devolver todo cero
    if len(idx_keep) == 0:
        return np.zeros(N_total)

    # Normalizar fracciones de los que quedan
    y_keep = y[idx_keep]
    y_keep /= y_keep.sum()

    # Funciones e info reducida a los que quedan
    iso_fun_keep = [iso_fun[i] for i in idx_keep]

    # --- cálculo según método ---
    if method.lower() == "naive":
        q_keep = np.array([iso_fun_keep[i](y_keep[i] * Ptot_kPa, T) for i in range(len(y_keep))])

    elif method.lower() == "iast":
        from isoLibs import IAST
        P_i_kPa = y_keep * Ptot_kPa
        q_keep = IAST(iso_fun_keep, P_i_kPa, T)

    elif method.lower() == "rast":
        from isoLibs import rast
        if Lamb is None:
            raise ValueError("RAST requiere matriz Lamb NxN")
        # Reducir Lamb a las especies mantenidas
        Lamb_keep = np.array(Lamb)[np.ix_(idx_keep, idx_keep)]
        P_i_kPa = y_keep * Ptot_kPa
        q_keep = rast(iso_fun_keep, P_i_kPa, T, Lamb_keep, C)

    else:
        raise ValueError("Método no reconocido")

    # --- reconstruir vector q completo ---
    q_full = np.zeros(N_total)
    q_full[idx_keep] = q_keep
    q_full[idx_drop] = 0.0
    return q_full
```

```python
def _save_isoFuncs(isoFuncs, filename):
    """
    Guarda IsoFuncs en un archivo JSON (sin las funciones directas).
    """
    data_to_save = {
        'species': isoFuncs['species'],
        'iso_fun_name': isoFuncs['iso_fun_name'],
        'iso_fun_par': [list(map(float, p)) for p in isoFuncs['iso_fun_par']],
        'iso_qs0': list(map(float, isoFuncs['iso_qs0'])),
        'iso_dH': [None if v is None else float(v) for v in isoFuncs['iso_dH']],
        'iso_Tref': [None if v is None else float(v) for v in isoFuncs['iso_Tref']],
        'err_mse': list(map(float, isoFuncs['err_mse'])),
    }
    with open(filename, 'w') as f:
        json.dump(data_to_save, f, indent=2)
    print(f"[OK] IsoFuncs guardado en {filename}")
```

```python
def _load_isoFuncs(filename):
    """
    Lee IsoFuncs desde archivo JSON y reconstruye las funciones q(P,T).
    """
    with open(filename, 'r') as f:
        data = json.load(f)

    model_map = {
        'Langmuir': Lang,
        'Freundlich': Freu,
        'Quadratic': Quad,
        'Sips': Sips,
        'Dual-site Langmuir': DSLa,
        'Dual-site Langmuire': DSLa  # por si acaso
    }

    fun_list = []
    for name, pars, dH, Tref in zip(data['iso_fun_name'],
                                    data['iso_fun_par'],
                                    data['iso_dH'],
                                    data['iso_Tref']):
        base_fn = model_map[name]

        if dH is None or Tref is None:
            iso_fun_PT = lambda P, T, base=base_fn, pars=pars: base(pars, P)
        else:
            iso_fun_PT = iso2isoArr(lambda P: base(pars, P), dH, Tref)

        fun_list.append(iso_fun_PT)

    isoFuncs = {
        'species': data['species'],
        'iso_fun': fun_list,
        'iso_fun_name': data['iso_fun_name'],
        'iso_fun_par': [np.array(p) for p in data['iso_fun_par']],
        'iso_qs0': np.array(data['iso_qs0']),
        'iso_dH': data['iso_dH'],
        'iso_Tref': data['iso_Tref'],
        'err_mse': data['err_mse'],
    }
    print(f"[OK] IsoFuncs cargado desde {filename}")
    return isoFuncs
```

```python
import json, zlib, base64

# ---------- SERIALIZAR ----------
def _pack_isoFuncs(isoFuncs: dict) -> str:
    """
    Convierte IsoFuncs en una cadena base64 (JSON comprimido con zlib).
    No serializa lambdas; guarda nombre de modelo, parámetros, dH, Tref, etc.
    """
    payload = {
        "schema": "IsoFuncs.v1",
        "species": isoFuncs["species"],
        "iso_fun_name": isoFuncs["iso_fun_name"],
        "iso_fun_par": [list(map(float, p)) for p in isoFuncs["iso_fun_par"]],
        "iso_qs0": list(map(float, np.asarray(isoFuncs["iso_qs0"]).ravel())),
        "iso_dH": [None if v is None else float(v) for v in isoFuncs["iso_dH"]],
        "iso_Tref": [None if v is None else float(v) for v in isoFuncs["iso_Tref"]],
        "err_mse": list(map(float, isoFuncs.get("err_mse", [np.nan]*len(isoFuncs["species"])))),
    }
    raw = json.dumps(payload, separators=(",", ":")).encode("utf-8")
    comp = zlib.compress(raw, level=9)
    b64  = base64.b64encode(comp).decode("ascii")
    return b64

# ---------- DESERIALIZAR ----------
def _unpack_isoFuncs(encoded_str: str) -> dict:
    """
    Decodifica la cadena base64 y reconstruye IsoFuncs, incluyendo las funciones q(P,T).
    """
    model_map = {
        "Langmuir": Lang,
        "Freundlich": Freu,
        "Quadratic": Quad,
        "Sips": Sips,
        "Dual-site Langmuir": DSLa,
        "Dual-site Langmuire": DSLa,  # tolerancia a nombre
    }

    comp = base64.b64decode(encoded_str.encode("ascii"))
    raw  = zlib.decompress(comp)
    data = json.loads(raw.decode("utf-8"))

    if data.get("schema") != "IsoFuncs.v1":
        raise ValueError(f"Esquema no soportado: {data.get('schema')}")

    species      = data["species"]
    fun_names    = data["iso_fun_name"]
    fun_pars     = data["iso_fun_par"]
    qs0          = np.array(data["iso_qs0"], dtype=float)
    dH_list      = data["iso_dH"]
    Tref_list    = data["iso_Tref"]
    err_mse_list = data.get("err_mse", [np.nan]*len(species))

    fun_list = []
    for name, pars, dH, Tref in zip(fun_names, fun_pars, dH_list, Tref_list):
        if name not in model_map:
            raise ValueError(f"Modelo no implementado: {name}")
        base_fn = model_map[name]
        pars_np = np.array(pars, dtype=float)

        # Si hay dependencia con T (dH y Tref), usa iso2isoArr; si no, solo P.
        if dH is None or Tref is None:
            iso_fun_PT = (lambda P, T, base=base_fn, pp=pars_np: base(pp, P))
        else:
            iso_fun_PT = iso2isoArr(lambda P, base=base_fn, pp=pars_np: base(pp, P),
                                    dH=float(dH), T_ref=float(Tref))

        fun_list.append(iso_fun_PT)

    isoFuncs = {
        "species": species,
        "iso_fun": fun_list,             # funciones q(P,T) reconstruidas
        "iso_fun_name": fun_names,
        "iso_fun_par": [np.array(p) for p in fun_pars],
        "iso_qs0": qs0,
        "iso_dH": dH_list,
        "iso_Tref": Tref_list,
        "err_mse": err_mse_list,
    }
    return isoFuncs
```
