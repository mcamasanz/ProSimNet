```python
import sys, os
import numpy as np

ruta_libs = os.path.abspath("../libs")
if ruta_libs not in sys.path:
    sys.path.append(ruta_libs)

from materialLibs import *
from unitsLibs import Tank, Valve
from solveLibs import solveTankWithValves
```

## LLENADO DE TANQUE CON APERTURA LINEAL Y LOGICA LINEAL

```python
# 1. Definir especies y propiedades
species = ["CO2", "N2"]
WM, mu, Dm, cpg_molar, cpg_mass, k, H = get_properties_gas(species)
propiedades_gas = {
    "species": species,
    "MW": WM,
    "mu": mu,
    "Dm": Dm,
    "Cp_molar": cpg_molar,
    "Cp_mass": cpg_mass,
    "k": k,
    "H": H}
```

```python
# 2. Configuracion de los equipos y condiciones iniciles y operacion
valve_inlet = Valve(
    Name="Entrada [Lin-Step-CO]",
    Cv_max= 1.35,  
    valve_type="linear",
    a_min = 0.,
    a_max = 0.6,
    logic= "linear",
    logic_params={"start": 1, "duration": 15},
    opening_direction="co")

valve_outlet = Valve(
    Name="Salida [Lin-Step-CO]",
    Cv_max= 1.35,  # cerrada
    valve_type="linear",
    a_min = 0.,
    a_max = 1.,
    logic= "linear",
    logic_params={"start": 60, "duration": 1e-6},
    opening_direction="co")

tank = Tank(
    Name="Tanque01",
    Longitud=2.0,     # m
    Diametro=.61,     # m
    Espesor=0.02,     # m
    prop_gas=propiedades_gas,
)

# 3. Condiciones iniciales tanque
tank.initialC_info(P0=3.5e5,T0=298.0,x0=[0.5, 0.5])
# 4. Condiciones de frontera (alimentaci√≥n)
tank.boundaryC_info(Pin=9e5,Tin=350.0,xin=[0.2, 0.8],Pout=1e5)
# 5. Thermal_info
tank.thermal_info(adi=True,kw=1e99,hint=1e99,hext=1e99,Tamb=298.15)
# 7. Lanzar
solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.1,
    startTime=0,
    endTime=12,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True,
    logBalances=True,
    )
```

```python
#croopTime inicializar con 0 segundos
solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.1,
    startTime=0,
    endTime=12,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True,
    logBalances=True,
    )

#croopTime retroceder a 6 segundos
solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.1,
    startTime=6,
    endTime=12,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True,
    logBalances=True,
    )
#croopTime empezar desde el ultimo tiempo guardado
solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.1,
    startTime=6,
    endTime=24,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True,
    logBalances=True,
    )
```

### LLENADO DE TANQUE CON APERTURA LINEAL Y LOGICA ESCALONADA

```python
# 6. Cambiar configuracion de la valvula de entrada
config_new_inlet = {
    "Name": "Entrada [Lin-Step-OC]",            # Nombre de la v√°lvula (descriptivo, sin efecto en l√≥gica)
    "Cv_max": 1.35,                             # Coeficiente m√°ximo de caudal [Nm¬≥/h¬∑bar^0.5]
    "valve_type": "linear",                     # Tipo de v√°lvula: Cv aumenta proporcionalmente con la apertura (lineal)
    "a_min": 0.,
    "a_max": 1.,
    "logic": "step",                            # L√≥gica: apertura escalonada en pasos discretos
    "logic_params": {
        "start": 5,                             # Inicio de la l√≥gica de apertura [s]
        "duration": 1e-6,                       # Duraci√≥n total hasta alcanzar el valor m√°ximo de apertura [s]
        "steps": 1                              # N√∫mero total de pasos (cada paso aumenta/disminuye 10‚ÄØ%)
    },
    "opening_direction": "co"                   # Direcci√≥n 'co' (close-to-open): empieza completamente abierta (100‚ÄØ%) y se va cerrando escalonadamente
}

valve_inlet.update(config_new_inlet)          

solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=1,
    startTime=0,
    endTime=10,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True)
```

```python
# 6. Cambiar configuracion de la valvula de entrada
config_new_inlet = {
    "Name": "Entrada [Lin-Step-OC]",            # Nombre de la v√°lvula (descriptivo, sin efecto en l√≥gica)
    "Cv_max": 1.35,                             # Coeficiente m√°ximo de caudal [Nm¬≥/h¬∑bar^0.5]
    "valve_type": "linear",                     # Tipo de v√°lvula: Cv aumenta proporcionalmente con la apertura (lineal)
    "a_min": 0.,
    "a_max": 1.,
    "logic": "step",                            # L√≥gica: apertura escalonada en pasos discretos
    "logic_params": {
        "start": 2,                             # Inicio de la l√≥gica de apertura [s]
        "duration": 8,                       # Duraci√≥n total hasta alcanzar el valor m√°ximo de apertura [s]
        "steps": 5                              # N√∫mero total de pasos (cada paso aumenta/disminuye 10‚ÄØ%)
    },
    "opening_direction": "co"                   # Direcci√≥n 'co' (close-to-open): empieza completamente abierta (100‚ÄØ%) y se va cerrando escalonadamente
}

valve_inlet.update(config_new_inlet)           #update hace el rest de forma implicita del valvula

solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.5,   
    startTime=0,
    endTime=10,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True)
```

### LLENADO DE TANQUE CON APERTURA LINEAL Y LOGICA SENOIDAL

```python
config_new_inlet ={
    "Name": "Lineal-Sin-OC",                    # Nombre descriptivo de la v√°lvula
    "Cv_max": 1.35,                             # Coeficiente m√°ximo de caudal [Nm¬≥/h¬∑bar^0.5]
    "valve_type": "linear",                      # Tipo de v√°lvula: Cv var√≠a linealmente con la apertura (Cv = a * Cv_max)
    "a_min": 0.,
    "a_max": 1.,
    "logic": "sin",                              # L√≥gica: variaci√≥n sinusoidal de la apertura
    "logic_params": {
        "start": 1,                              # Tiempo inicial de acci√≥n [s]
        "duration": 8,                          # Tiempo total del ciclo de apertura [s]
        "freq": 10                                # N√∫mero de ciclos completos (oscilaciones) dentro del intervalo de duraci√≥n
    },
    "opening_direction": "co"                   # Direcci√≥n 'co' (close-to-open): se interpreta como inverso del perfil sinusoidal
}

valve_inlet.update(config_new_inlet)           #update hace el rest de forma implicita del valvula

solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.5,
    startTime=0,
    endTime=10,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True)
```

## LLENADO DE TANQUE CON APERTURA LINEAL Y LOGICA SIGMOIDAL

```python
config_new_inlet= {
    "Name": "Lineal-Sigmoid-OC",                # Nombre descriptivo de la v√°lvula
    "Cv_max": 1.35,                             # Coeficiente m√°ximo de caudal [Nm¬≥/h¬∑bar^0.5]
    "valve_type": "linear",                      # Tipo de v√°lvula: Cv escala linealmente con la apertura (Cv = a * Cv_max)
    "logic": "sigmoid",                          # L√≥gica de control: curva sigmoide (suave al inicio y al final, m√°s r√°pida al centro)
    "logic_params": {
        "start": 2,                              # Tiempo en que empieza a ejecutarse la l√≥gica [s]
        "duration": 5,                          # Tiempo total para completar el perfil de apertura [s]
        "k": 10,                                 # Pendiente o "agudeza" de la curva (m√°s alto = m√°s brusco)
        "x0": 0.5                                # Punto de inflexi√≥n (normalizado entre 0 y 1 ‚Üí a la mitad del tiempo)
    },
    "opening_direction": "co"                   # Direcci√≥n 'co' (close-to-open): se invierte la curva (empieza abierta y cierra suavemente)
}

valve_inlet.update(config_new_inlet)           #update hace el rest de forma implicita del valvula

solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.5,
    startTime=0,
    endTime=10,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True)
```

## LLENADO DE TANQUE CON APERTURA LINEAL Y LOGICA POLINIMICA

```python
config_new_inlet= {
    "Name": "Custom-Poly",                       # Nombre de la v√°lvula (identificativo para gr√°ficos o seguimiento)
    "Cv_max": 1.35,                             # Caudal m√°ximo en condiciones normales [Nm¬≥/h¬∑bar^0.5]
    "valve_type": "linear",                      # Relaci√≥n lineal entre apertura y coeficiente Cv (Cv = a * Cv_max)
    "a_min": 0.,
    "a_max": 1.,
    "logic": "poly",                             # L√≥gica: apertura basada en una funci√≥n polin√≥mica (potencia)
    "logic_params": {
        "start": 1,                              # Momento inicial del ciclo de apertura [s]
        "duration": 5,                          # Tiempo total hasta alcanzar apertura completa [s]
        "degree": 2                              # Grado del polinomio: cuanto mayor, m√°s lenta la apertura al inicio
    },
    "opening_direction": "co"                    # Direcci√≥n: 'close-to-open' ‚Üí comienza abierta (100‚ÄØ%) y se cierra
}

valve_inlet.update(config_new_inlet)           #update hace el rest de forma implicita del valvula

solveTankWithValves(
    Tank=tank,
    Valve_inlet=valve_inlet,
    Valve_outlet=valve_outlet,
    saveData=0.5,
    endTime=10,
    solver='BDF',
    atol=1e-12,
    rtol=1e-12,
    plot=True)
```
