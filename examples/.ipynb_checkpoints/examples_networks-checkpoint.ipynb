```python
import sys, os
import numpy as np
from pprint import pprint


ruta_libs = os.path.abspath("../libs")
if ruta_libs not in sys.path:
    sys.path.append(ruta_libs)

from materialLibs import *
from unitsLibs import *
from solveLibs import *
from networkLibs import *
```

```python
species = ["CO2", "N2"]
WM, mu, Dm, cpg_molar, cpg_mass, k, H = get_properties_gas(species)
prop_gas = {
    "species": species,
    "MW": WM,
    "mu": mu,
    "Dm": Dm,
    "Cp_molar": cpg_molar,
    "Cp_mass": cpg_mass,
    "k": k,
    "H": H}
```

```python
tank_1 = Tank(
    Name="T1",
    Longitud=2.0,     # m
    Diametro=.61,     # m
    Espesor=0.02,     # m
    prop_gas=prop_gas)

vT1bi = Valve(
    Name="vT1bi",
    Cv_max= 1.35,  
    valve_type="linear",
    a_min = 0.,
    a_max = 1.,
    logic= "linear",
    logic_params={"start": 1, "duration": 15},
    opening_direction="co")

vT1to = Valve(
    Name="vT1to",
    Cv_max= 1.35,  
    valve_type="linear",
    a_min = 0.,
    a_max = 1.,
    logic= "linear",
    logic_params={"start": 1, "duration": 15},
    opening_direction="co")

tank_2 = Tank(
    Name="T2",
    Longitud=2.0,     # m
    Diametro=.61,     # m
    Espesor=0.02,     # m
    prop_gas=prop_gas)

vT2bi = Valve(
    Name="vT2bi",
    Cv_max= 1.35,  
    valve_type="linear",
    a_min = 0.,
    a_max = 1.,
    logic= "linear",
    logic_params={"start": 1, "duration": 15},
    opening_direction="co")

vT2to = Valve(
    Name="vT2to",
    Cv_max= 1.35,  
    valve_type="linear",
    a_min = 0.,
    a_max = 1.,
    logic= "linear",
    logic_params={"start": 1, "duration": 15},
    opening_direction="co")

vT1bT2b = Valve(
    Name="vT1bT2b",
    Cv_max= 1.35,  
    valve_type="linear",
    a_min = 0.,
    a_max = 1.,
    logic= "linear",
    logic_params={"start": 1, "duration": 15},
    opening_direction="co")

vT1tT2t = Valve(
    Name="vT1tT2t",
    Cv_max= 1.35,  
    valve_type="linear",
    a_min = 0.,
    a_max = 1.,
    logic= "linear",
    logic_params={"start": 1, "duration": 15},
    opening_direction="co")

valves=[vT1bi,vT1to,vT2bi,vT2to,vT1tT2t,vT1bT2b]
units=[tank_1,tank_2,]
```

```python
sim=Network(prop_gas=prop_gas,
            Units=units,
            Valves=valves)
```

```python
sim._updateNetwork()
sim._printNetwork()
```
